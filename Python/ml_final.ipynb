```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()

# importing the dataset
df  = pd.read_csv("Iris.csv")
df.head()
```

```python
df = df.drop(["Id"],axis=1)
df.head(100)
```

```python
df["Species"] = le.fit_transform(df["Species"])
df.head()
```

```python

df.info()
df.head(75)
```

Adding the calsifiaction col 

```python
b = []
for i in df["Species"]:
    if i == 1:
        b.append("Iris-versicolor")
    else:
        b.append("Non - Iris-versicolor")
df["claasName"] = b

print(df)
df.head(100)
```

changing clasfication to a number 

```python
df["claasName"] = le.fit_transform(df["claasName"])

df = df.drop(["Species"],  axis = 1)

df.info()

df.head(100)
```

Spillting the data

```python
X = df.iloc[:,[0,1,2,3]].values
y = df.iloc[:,4].values

X = X.reshape(-1, 4)
X = preprocessing.scale(X) #scale the data so that it is easier to fit

X_train, X_test, y_train, y_test = train_test_split(X, 
                y, test_size=0.25, random_state=0)
```

```python
# fit a model
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# make predictions on test data
y_pred = log_reg.predict(X_test).reshape(-1,1)
```

```python
# Use score method to get accuracy of model
score = log_reg.score(X_test, y_test)

print('Accuracy: {}'.format(score))
```

The confusion matrix 

```python
from sklearn.metrics import confusion_matrix

classes = ["Iris-versicolor ","Non - Iris-versicolor"]
conf_mat = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(conf_mat, columns=classes, index=classes)
cm_df
```

from the confution matrix we can exspect that there would be a high recall value and a slightly above avrage accuracy rating 

```python
from sklearn.metrics import f1_score, precision_score, recall_score

# average f1 score
av_f1 = f1_score(y_test, y_pred, average='micro')
print(av_f1)

# f1 score per class
f = f1_score(y_test, y_pred, average=None)
lowest_score = min(f)
hardest_class = classes[list(f).index(lowest_score)]
print('Hardest class:', hardest_class)
```

Score report 

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

Here we cans see that that recall value for iris sorta is 0.95 and the pricition is 0.64.
